BEGINNER LEVEL TASK: TASK 4 Image to Pencil Sketch with Python

```python
import cv2
from google.colab.patches import cv2_imshow
```

```python
#read the image
image = cv2.imread("cat.PNG")
cv2_imshow( image)
cv2.waitKey(0)
```

```python
#creat a new image by converting the original image to greyscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_image)
cv2.waitKey(0)
```

```python
#inverting the new grayscale image
inverted_image = 255 - gray_image
cv2_imshow(inverted_image)
cv2.waitKey()
```

```python
#blur the image by using the Gaussian Function in OpenCV
blurred = cv2.GaussianBlur(inverted_image, (21, 21), 0)
```

```python
#invert the blurred image
inverted_blurred = 255 - blurred
pencil_sketch = cv2.divide(gray_image, inverted_blurred, scale=256.0)
cv2_imshow(pencil_sketch)
cv2.waitKey(0)
```

```python
cv2_imshow(image)
cv2_imshow(pencil_sketch)
cv2.waitKey(0)
```
